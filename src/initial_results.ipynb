```python
def print_mean_summary(results, mean, index):
    print("Means for the six month forward rate")
    print('This is the mean of the results model one' , np.mean(results[0][:,index]))
    print('This is the mean of the results model two' , np.mean(results[1][:,index]))
    print('This is the mean of the results model three', np.mean(results[2][:,index]))
    print('This is the mean of the results model four', np.mean(results[3][:,index]))
    print('This is the mean of the results model five', np.mean(results[4][:,index]))
    print('This is the actual mean of the forwards', np.mean(X[:,index]))
                  
```
